```python
#this script read the sigma_est and sigma_exact estimated for each test map and plot one versus the other
import sys
import h5py
import numpy as np
import healpy as hp
import random as python_random
import matplotlib.pyplot as plt
import pandas as pd
import os, shutil
```

```python
home_dir="/home/amorelli/r_estimate/B_maps_white_noise/results_16_5_23/"
r_like=[0.0,0.001,0.005,0.007,0.01] #these are the r_test of the maps on which i test the NN 
f_sigma_like=np.load(home_dir+"sigma_exact.npz")
map_names=f_sigma_like.files
sigma_like=[]
for name in map_names:
    sigma_like.append(f_sigma_like[name])
```

```python
f_sigma_est=np.load(home_dir+"sigma_out.npz")
sigma_est=[]
for name in map_names:
    sigma_est.append(f_sigma_est[name])
```

```python
dim_fig=100
half_dim=int(len(map_names)/2)
for i in range(1,half_dim+2):
    if i*i>=len(map_names):
        dim_fig=i
        break
fig = plt.figure(figsize=(20,20))
for i,name in enumerate(map_names):
    plt.subplot(dim_fig,dim_fig,i+1)
    plt.plot(sigma_est[i],sigma_like[i], color='r',linestyle='None',marker='.', markersize = 10.0)
    plt.title(name)
    plt.ylabel("sigma_like")
    plt.xlabel("sigma_est")
```
